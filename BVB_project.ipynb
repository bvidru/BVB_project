### Create DataFrames for every each sheet used for analysis

```python
# create dataframes for every sheet from all files used for analysis

def concatenate_dataframes_with_same_key(dict_list, key):
    """
    Concatenate DataFrames with the same key from a list of dictionaries
    and return a new DataFrame.

    Args:
    dict_list (list): List of dictionaries where values are DataFrames.
    key (str): The key to identify DataFrames with the same key.

    Returns:
    pd.DataFrame: A concatenated DataFrame.
    """
    # Create a dictionary to group DataFrames by their key
    grouped_dataframes = {}
    for d in dict_list:
        if key in d and isinstance(d[key], pd.DataFrame):
            if key not in grouped_dataframes:
                grouped_dataframes[key] = []
            grouped_dataframes[key].append(d[key])

    # Concatenate DataFrames within each group
    concatenated_dataframes = []
    for df_list in grouped_dataframes.values():
        if len(df_list) > 1:
            concatenated_df = pd.concat(df_list, ignore_index=True)
            concatenated_dataframes.append(concatenated_df)

    # Concatenate the DataFrames from all groups into a single DataFrame
    result_df = pd.concat(concatenated_dataframes, ignore_index=True)

    return result_df
```

In order to assign the date on every obs, in cronological order, from each file extracted, I will create a new column named ['Groups'] in which I will divide the bonds into two groups: corporate bonds and institutional bonds

```python
# Define a function to assign group id's based on the starting letter
def assign_id(name):
    if str(name).startswith('R2'):
        return 'IB'
    else:
        return 'CB'

# Apply the function to the DataFrame column
df_bonds['ID'] = df_bonds['Symbol'].apply(assign_id)
df_bonds
```

```python
# create a date column and assign the same date to all observations in a group. The date will increase by one for every group.

# Create a new column `Group` in `df` to store the group labels
df_bonds['Group'] = ''

# Create a dictionary to map the group labels to the corresponding rows in `df`
group_map = {}

# Iterate over the rows of `df` and assign the group label to the `Group` column and the group map
for i in range(len(df_bonds)):
    if df_bonds.loc[i, 'ID'] == 'CB':
      if i > 0 and df_bonds.loc[i-1, 'ID'] == 'CB':
        df_bonds.loc[i, 'Group'] = df_bonds.loc[i - 1, 'Group']
      else:
        group_label = str(len(group_map)+1)
        df_bonds.loc[i, 'Group'] = group_label
        group_map[group_label] = [i]
    elif df_bonds.loc[i, 'ID'] == 'IB':
        if df_bonds.loc[i - 1, 'ID'] == 'CB':
            group_label = df_bonds.loc[i - 1, 'Group']
            df_bonds.loc[i, 'Group'] = group_label
            group_map[group_label] = [i]
        else:
            group_label = df_bonds.loc[i - 1, 'Group']
            df_bonds.loc[i, 'Group'] = group_label
            group_map[group_label].append(i)

# Create a new column 'Date' with assigned dates
date_dict = {}
current_date = pd.Timestamp('2023-01-01')

df_bonds['Date'] = df_bonds['Group'].apply(lambda x: date_dict.setdefault(x, current_date + pd.DateOffset(days=len(date_dict))))
# Print the new DataFrame `data1`
df_bonds
```
